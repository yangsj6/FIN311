# 1. Import data

```python
import dataset
from matplotlib import pyplot as plt
xs,ys = dataset.get_beans(100)
```

# 2. Visualize data

```python
plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")
plt.scatter(xs,ys)
plt.show();
```

# Step 3: Train the model

```python
w = 0.5

for m in range(100):
    for i in range(100):
        x = xs[i]
        y = ys[i]
        y_pre = w * x
        e = y - y_pre
        alpha = 0.05
        w = w + alpha*e*x

y_pre = w * xs

plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")
plt.scatter(xs,ys)
plt.plot(xs,y_pre,"r")
plt.show();
```
